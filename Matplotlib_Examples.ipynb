# Matplotlib â€” Data Visualization in Python

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a simple line plot
plt.figure(figsize=(8, 4)) # Set figure size
plt.plot(x, y, label='sin(x)', color='blue', linestyle='--')
plt.title('Simple Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.show() # Display the plot
```

### Example: Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for scatter plot
np.random.seed(42) # for reproducibility
x_scatter = np.random.rand(50) * 10
y_scatter = np.random.rand(50) * 10 + x_scatter * 0.5 # Add some correlation

# Create a scatter plot
plt.figure(figsize=(8, 4))
plt.scatter(x_scatter, y_scatter, color='red', marker='o', alpha=0.7, label='Random Data')
plt.title('Scatter Plot of Random Data')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()
```

### Example: Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 12, 39]

# Create a bar chart
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum'])
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales (Units)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Example: Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for histogram (e.g., heights)
data_hist = np.random.normal(loc=170, scale=10, size=1000) # Mean 170, Std Dev 10

# Create a histogram
plt.figure(figsize=(8, 5))
plt.hist(data_hist, bins=30, color='teal', edgecolor='black', alpha=0.7)
plt.title('Distribution of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Example: Multiple Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x_multi = np.linspace(0, 2 * np.pi, 100)
y1_multi = np.sin(x_multi)
y2_multi = np.cos(x_multi)
y3_multi = np.tan(x_multi) # Will have asymptotes
y4_multi = np.exp(-x_multi/2) * np.sin(2*x_multi)

# Create a figure with 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Multiple Subplots Example', fontsize=16)

# Plot 1: Sine wave
axes[0, 0].plot(x_multi, y1_multi, color='blue')
axes[0, 0].set_title('Sine Wave')
axes[0, 0].grid(True)

# Plot 2: Cosine wave
axes[0, 1].plot(x_multi, y2_multi, color='green')
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].grid(True)

# Plot 3: Tangent wave (with limits to avoid asymptotes)
axes[1, 0].plot(x_multi, y3_multi, color='red')
axes[1, 0].set_title('Tangent Wave')
axes[1, 0].set_ylim(-5, 5) # Limit y-axis for better visualization
axes[1, 0].grid(True)

# Plot 4: Damped Sine wave
axes[1, 1].plot(x_multi, y4_multi, color='purple')
axes[1, 1].set_title('Damped Sine Wave')
axes[1, 1].grid(True)

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
plt.show()
```

### Example: Customizing a Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x_custom = np.linspace(0, 5, 50)
y_custom = x_custom**2

# Create a plot with custom styles
plt.figure(figsize=(9, 6))
plt.plot(x_custom, y_custom,
         color='#FF5733',       # Hex color
         linestyle='-',         # Solid line
         linewidth=3,           # Line thickness
         marker='o',            # Circle markers
         markersize=8,          # Marker size
         markeredgecolor='black', # Marker border color
         markerfacecolor='gold', # Marker fill color
         label='$y = x^2$')     # LaTeX-style label

plt.title('Customized Parabola Plot', fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel('X-axis', fontsize=14, color='gray')
plt.ylabel('Y-axis', fontsize=14, color='gray')

plt.grid(True, linestyle=':', alpha=0.8)
plt.legend(fontsize=12, loc='upper left')

# Add an annotation
plt.annotate('Local Minimum', xy=(0, 0), xytext=(1, 10),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='green')

plt.show()
```

### Example: 3D Surface Plot

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D # Required for 3D plots

# Generate data for a 3D surface
x_3d = np.linspace(-5, 5, 100)
y_3d = np.linspace(-5, 5, 100)
X_3d, Y_3d = np.meshgrid(x_3d, y_3d)
Z_3d = np.sin(np.sqrt(X_3d**2 + Y_3d**2))

# Create a 3D figure
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X_3d, Y_3d, Z_3d, cmap='viridis', edgecolor='none')

ax.set_title('3D Surface Plot of sin(sqrt(x^2 + y^2))')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.show()
```
