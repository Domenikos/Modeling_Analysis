### Signal Processing

#### Low-pass Filtering a Signal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

# Generate a noisy signal
fs = 1000       # Sampling frequency
f1 = 10        # Frequency of the first sine wave
f2 = 50        # Frequency of the second sine wave
# Create a time vector of 1 second
t = np.linspace(0, 1, fs, endpoint=False) # Time vector
clean_signal = 0.5 * np.sin(2 * np.pi * f1 * t) + 0.3 * np.sin(2 * np.pi * f2 * t)
noise = 0.1 * np.random.randn(len(t))
noisy_signal = clean_signal + noise

# Design a Butterworth low-pass filter
# Cutoff frequency is 20 Hz (normalized to Nyquist frequency)
cutoff_freq = 20
nyquist_freq = 0.5 * fs
normal_cutoff = cutoff_freq / nyquist_freq
order = 4 # Filter order

b, a = butter(order, normal_cutoff, btype='low', analog=False)

# Apply the filter to the noisy signal
filtered_signal = lfilter(b, a, noisy_signal)

# Plot the original and filtered signals
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, noisy_signal, label='Noisy Signal', color='red', alpha=0.5)
plt.plot(t, clean_signal, label='Clean Signal', color='blue')
plt.title('Noisy Signal vs Clean Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.subplot(2, 1, 2)
plt.plot(t, filtered_signal, label='Filtered Signal', color='green')
plt.title(f'Filtered Signal (Butterworth Low-pass, $f_c$ = {cutoff_freq} Hz)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# Display results
print("\n--- scipy.signal Example (Filtering) ---")
print(f"Original signal length: {len(noisy_signal)}")
print(f"Filtered signal length: {len(filtered_signal)}")
print(f"Filtering applied: 4th order Butterworth low-pass filter with {cutoff_freq} Hz cutoff.")
print("-" * 40)
```

### Interpolation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d


# Original data points
x_data = np.array([0, 1, 2, 3, 4, 5])
y_data = np.array([0, 0.8, 0.9, 0.1, -0.8, -1.0])

# Create an interpolation function (cubic spline)
f_interp = interp1d(x_data, y_data, kind='cubic')

# New points where we want to interpolate
x_new = np.linspace(0, 5, num=50, endpoint=True)

# Get the interpolated y values
y_interpolated = f_interp(x_new)

# Plotting the original data and the interpolated curve
plt.figure(figsize=(10, 6))
plt.plot(x_data, y_data, 'o', label='Original Data', markersize=8)
plt.plot(x_new, y_interpolated, '-', label='Cubic Spline Interpolation', linewidth=2)
plt.title('Cubic Spline Interpolation Example')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()

print("\n--- scipy.interpolate Example (1-D Interpolation) ---")
print(f"Original x data: {x_data}")
print(f"Original y data: {y_data}")
print(f"Interpolated y at x=0.5: {f_interp(0.5):.4f}")
print(f"Interpolated y at x=2.5: {f_interp(2.5):.4f}")
# In a real application, you would plot original and interpolated points
# to visualize the spline.
print("-" * 40)
```
